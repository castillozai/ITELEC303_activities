```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Family Income and Expenditure.csv")
```

```python
data.head()
```

```python
data.describe(include="all")
```

```python
plt.figure(figsize=(10,5))
data.groupby("Region")["Total Household Income"].mean().plot(kind="bar")
plt.title("Average Income per Region")
plt.ylabel("Average Income")
plt.xlabel("Region")
plt.show()
```

<h1>INSIGHT: 1. Ranks regions by financial capacity. Useful for identifying wealthiest vs. most economically challenged areas. De</h1>
<h1>2. The NCR (National Capital Region) has the highest average household income by a considerable margin, exceeding ₱400,000.</h1>
<h1>3. The regions with the next highest average incomes are III - Central Luzon and IVA - CALABARZON, both close to or above ₱300,000. These three regions (NCR, CALABARZON, Central Luzon) are often considered the economic powerhouses of the Philippines.</h1>

```python
group = data.groupby("Region")[["Total Household Income", "Total Food Expenditure"]].mean()

plt.figure(figsize=(10,6))
group.plot(kind='bar')
plt.title("Average Income vs Expenditure per Region")
plt.xlabel("Region")
plt.ylabel("Amount")
plt.show()
```

<h1>INSIGHT:1. Compares earning and spending side-by-side. Regions where expenditure is close to income may indicate lower savings or high living cost.P </h1>
<h1>2. The NCR is the clear outlier with the highest income, while food expenditure remains relatively consistent across most regions, meaning lower-income regions spend a much larger percentage of their earnings on food compared to the NCR.</h1>

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

X = data[['Total Household Income']]
y = data['Bread and Cereals Expenditure']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Scatter + Regression Line
plt.figure(figsize=(10,6))
plt.scatter(X, y)

# Fix: Create DataFrame with correct feature name
x_line = np.linspace(X.min()[0], X.max()[0], 200)
x_line_data = pd.DataFrame({'Total Household Income': x_line})
y_line = model.predict(x_line_data)

plt.plot(x_line, y_line)
plt.title("Predicting Bread & Cereal Spending From Income", fontsize=16)
plt.xlabel("Total Household Income")
plt.ylabel("Bread and Cereals Expenditure")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("R² Score:", r2_score(y_test, y_pred))
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

<h1>INSIGHT:1.The relationship between total household income and bread & cereals spending is very weak.
    
Even as income increases, spending on basic staples barely increases.
    The regression line slopes upward slightly, but the effect is minimal.
</h1>

<h1>2. Most households (≈90%) earn below 2,000,000 and cluster tightly around the same spending range.
    
    
High-income households (above 5,000,000–10,000,000) do not spend proportionally more—their food spending remains within similar ranges</h1>

```python
lux_cols = [
    'Restaurant and hotels Expenditure',
    'Alcoholic Beverages Expenditure',
    'Tobacco Expenditure'
]

avg_lux = data[lux_cols].mean()

plt.figure(figsize=(8,6))
plt.bar(avg_lux.index, avg_lux.values)
plt.title("Average Spending on Luxury Categories", fontsize=16)
plt.ylabel("Average Amount")
plt.xticks(rotation=40)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

<h1>INSIGHT:1. The luxury spending bar chart indicates that restaurant and hotel expenses dominate household luxury consumption, accounting for nearly 78% of the total among the three categories shown, which highlights a strong preference for leisure and dining experiences over alcohol or tobacco.</h1>

<h1>2. It reveals that alcohol and tobacco spending combined make up only a small portion of luxury expenditures, which may suggest that households view these items as lower-priority or occasional indulgences compared to dining out and leisure activities.</h1>

```python
import seaborn as sns
import matplotlib.pyplot as plt

expense_cols = ['Total Food Expenditure', 'Housing and water Expenditure', 'Transportation Expenditure',
                'Education Expenditure', 'Medical Care Expenditure', 'Total Household Income']

corr = data[expense_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlation of Expense Categories with Total HouseHold Income", fontsize=16)
plt.tight_layout()
plt.show()
```

<h1>INSIGHT:1. The correlation heatmap reveals that housing, food, and transportation expenses have the strongest positive relationships with income—ranging from 64% to 68%—indicating that as income grows, households allocate noticeably more money to these essential but flexible spending categories.</h1>

<h1>2. The heatmap shows that education expenditure has only a moderate correlation with income (44%), which implies that while higher-income families spend more on education, the increase is not as substantial as in other categories, possibly indicating that many households invest in education regardless of income constraints.</h1>
